# Data wrangling

```python
# Create a subset of the original data
"""I'm going to focus on the numbers of the persons injured or killed on an accident"""
subset = ['CRASH DATE', 'CRASH TIME', 'BOROUGH', 'ZIP CODE',
          'ON STREET NAME', 'CROSS STREET NAME',
        'NUMBER OF PERSONS INJURED',
       'NUMBER OF PERSONS KILLED', 
       'CONTRIBUTING FACTOR VEHICLE 1',
       'VEHICLE TYPE CODE 1'
         ]

df_subset = df[subset].copy()
df_subset.head()
```

```python
# Modify multiple columns names
df_subset.rename(columns = {'CRASH DATE':'DATE',
                    'CONTRIBUTING FACTOR VEHICLE 1':'CONTRIBUTING FACTOR VEHICLE',
                    'VEHICLE TYPE CODE 1':'VEHICLE TYPE CODE'}, inplace = True)
```

Change the structure of the data to the correct one

```python
# Check the type of the date column
df_subset['DATE'].dtype
```

```python
# It looks like the data is a object. Let's change it to a datetime
df_subset['DATE']= pd.to_datetime(df_subset['DATE'])
df_subset['DATE'].dtype
```

Now, let's fix the missing values

```python
# Let's check again the missing values
df_subset.info()
```

```python
# we can drop the CROSS STREET NAME  and filter the data to only include the rows that have BOROUGH OR ZIPCODE
df_subset.drop(columns = ['CROSS STREET NAME'], inplace = True)
# Filter the data to remove the NA's
df_clean = df_subset.dropna(subset=['BOROUGH']).copy()
```

Create new column based on existing columns

```python
# Let's create a new column that's true is collision led to death and false otherwise
df_clean['DEATH_FLAG'] = df_clean['NUMBER OF PERSONS KILLED'] > 0
df_clean.head()
```

Drop columns from the dataset:

Drop rows from the dataset

```python
# Drop the first row
df_clean.drop(index=0, inplace=True)
```

Sort the data based on multiple variables

```python
df_clean.sort_values(by=['BOROUGH', 'NUMBER OF PERSONS INJURED'], ascending=[True, False], inplace=True)
df_clean.head(5)
```

Convert string to lower case

```python
# We can convert the BOROUGH column in lower case
df_clean['BOROUGH'] = df['BOROUGH'].str.lower()
df_clean.head()
```

Check whether a numeric value is present in a given column

```python
# Let's check if the column BOROUGH has numeric value
pd.to_numeric(df_clean['BOROUGH'], errors='coerce').notna().sum()
```
